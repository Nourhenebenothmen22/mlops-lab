```python
import mlflow
import transformers
mlflow.set_tracking_uri('http://mlflow-web:5000')
mlflow.set_experiment("NLP_Transformers")

pipeline = transformers.pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
with mlflow.start_run():
    mlflow.transformers.log_model(pipeline, artifact_path="model", registered_model_name="DistilBERT_Sentiment")
    print("NLP Model Logged")
```
