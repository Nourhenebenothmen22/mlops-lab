```python
import mlflow
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes

mlflow.set_tracking_uri('http://mlflow-web:5000')
mlflow.set_experiment("experiment_diabetes")
mlflow.sklearn.autolog()

db = load_diabetes()
X_train, X_test, y_train, y_test = train_test_split(db.data, db.target, test_size=0.2)

with mlflow.start_run(run_name="rf_v1"):
    rf = RandomForestRegressor(n_estimators=100, max_depth=6)
    rf.fit(X_train, y_train)
    print("Run ID:", mlflow.active_run().info.run_id)
```
